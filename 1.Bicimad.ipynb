```python
#first, we import libraries, read all json files that are in the same folder
import os
import json 
import glob 
```

```python
# we merge the json files in two groups, from jan to june 2019 because they used ('$date') and the rest july 2019 to feb 2020

#Group 1 
os.chdir(r'C:\Users\Usuario\Documents\Daniela Arias\Tesis\Practical\Comparativo\Pre-covid\Bicimad\primero')
read_files_primero = glob.glob('*.json')

with open("merged_file1.json", "w") as outfile1:
    for f in read_files_primero:
        with open(f, "r") as infile1:
            outfile1.write(infile1.read())
```

```python
#group 2 from july to feb 2020 (sin '$date')
os.chdir(r'C:\Users\Usuario\Documents\Daniela Arias\Tesis\Practical\Comparativo\Pre-covid\Bicimad\segundo')
read_files_segundo = glob.glob('*.json')

with open("merged_file2.json", "w") as outfile2:
    for n in read_files_segundo:
        with open(n, "r") as infile2:
            outfile2.write(infile2.read())
```

```python
#now we read the merged files and clean them

os.chdir(r'C:\Users\Usuario\Documents\Daniela Arias\Tesis\Practical\Comparativo\Pre-covid\Bicimad\primero')
with open('merged_file1.json', 'r') as t:
    base1 = list(map(json.loads, t))
```

```python
os.chdir(r'C:\Users\Usuario\Documents\Daniela Arias\Tesis\Practical\Comparativo\Pre-covid\Bicimad\segundo')
with open('merged_file2.json', 'r') as z:
    base2 = list(map(json.loads, z))
```

```python
for d in base1:
    del d['_id']
    del d['user_day_code']
    del d['idplug_base']
    del d['idunplug_base']
    del d['ageRange']
    del d['zip_code']
```

```python
for d in base2:
    del d['_id']
    del d['user_day_code']
    del d['idplug_base']
    del d['idunplug_base']
    del d['ageRange']
    del d['zip_code']
```

```python
keys_to_keep = ['travel_time', 'idunplug_station', 'idplug_station','unplug_hourTime', 'user_type']

base11 = [{ key: item[key] for key in keys_to_keep } for item in base1]
base12 = [{ key: item[key] for key in keys_to_keep } for item in base2]
```
